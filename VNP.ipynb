````python
import re
````

````python
def extract_key_value_pairs_regex(text):
    key_value_pairs = {}
    features = []

    kv_pattern = re.compile(r'([\w\s\(\)]+):\s*(.*)')

    lines = text.split('\n')

    for i, line in enumerate(lines):
        match = kv_pattern.match(line)
        if match:
            key, value = match.groups()
            key_value_pairs[key.strip()] = value.strip()
        else:
            if i > 0:
                previous_line = lines[i-1]
                prev_match = kv_pattern.match(previous_line)
                if prev_match:
                    prev_key, prev_value = prev_match.groups()
                    if not prev_value.strip():
                        key_value_pairs[prev_key.strip()] = line.strip()
                    else:
                        features.append(line.strip())
                else:
                    features.append(line.strip())
            else:
                features.append(line.strip())

    if features:
        key_value_pairs['Features'] = ' '.join(features).strip()

    return key_value_pairs
````

````python
data['Key_Value_Pairs'] = data['Specifications'].apply(extract_key_value_pairs_regex)
````

````python
import cohere
import pandas as pd

# Initialize Cohere client
co = cohere.Client(COHERE_API_KEY)

# Function to extract key-value pairs for a single text
def extract_key_value_pairs_nlp(text):
    prompt = f"""
    Extract the key-value pairs from the following product specification text. If the key or value is missing, still infer the correct pairs. Identify additional information under the category 'Features'.

    Example 1:
    Input:
    "Screen Size: 6.5 inches
    Resolution: 1080x2400 pixels
    Features: HDR10, Dolby Vision
    Battery Capacity: 4000 mAh"

    Output:
    {{
        "Screen Size": "6.5 inches",
        "Resolution": "1080x2400 pixels",
        "Battery Capacity": "4000 mAh",
        "Features": "HDR10, Dolby Vision"
    }}

    Example 2:
    Input:
    "Processor: Intel i5
    RAM: 8GB
    Storage: 256GB SSD
    Display: 13.3 inches"

    Output:
    {{
        "Processor": "Intel i5",
        "RAM": "8GB",
        "Storage": "256GB SSD",
        "Display": "13.3 inches"
    }}

    Now, extract the key-value pairs for the following product specification text:

    {text}
    """

    response = co.generate(
        model='command-xlarge-nightly',
        prompt=prompt,
        max_tokens=200,
        temperature=0.3,
        stop_sequences=["}"]
    )

    output_text = response.generations[0].text.strip()

    try:
        output_dict = eval(output_text)
    except:
        output_dict = {"error": "Could not parse the response"}

    return output_dict


subset_data['extracted_features'] = subset_data['Specifications'].apply(extract_key_value_pairs_nlp)
````

**Advanced search**

````python
prompt_template = PromptTemplate(
    template="Extract the product features from the following query and provide them in JSON format. Please send in this format: {{\"Connection\": \"Bluetooth\", \"Power\": \"15W\"}}. Query: {query}. Please use my json format. DO not use [ ] or ```json ```. Do not say anything else.",
    input_variables=["query"]
)

cohere_llm = Cohere(cohere_api_key=COHERE_API_KEY)

cohere_chain = LLMChain(prompt=prompt_template, llm=cohere_llm)

def process_query_with_cohere(query):
    refined_features = cohere_chain.run(query=query)

    return refined_features

query = input("Enter search prompt:")
refined_features = process_query_with_cohere(query)

print("Refined Features using Cohere LLM:", refined_features)
````

**Translating keys of prompt**

````python
from translate import Translator

def translate_keys_translate(features, dest_lang='mk'):
    translator = Translator(to_lang=dest_lang)

    try:
      features = json.loads(features)
    except json.JSONDecodeError as e:
      print(f"Error parsing JSON: {e}")

    translated_features = {}
    print(type(features))
    if isinstance(features, dict):
        for key, value in features.items():
            if isinstance(key, str):
                try:
                    translated_key = translator.translate(key)
                except Exception as e:
                    print(f"Translation error for key '{key}': {e}")
                    translated_key = key
            else:
                print(f"Expected a string key, got: {type(key)}")
                translated_key = key

            translated_features[translated_key] = value
    else:
        print("Expected a dictionary, got:", type(features))

    return translated_features

translated_features = translate_keys_translate(refined_features)

print("Translated Features:", translated_features)
````

**Searching for translated keys in dataset (using first key-value-pairs column only)**

````python
import json

def search_compatible_items(df, search_prompt):
    search_dict = {}
    try:
        if isinstance(search_prompt, str):
            search_dict = json.loads(search_prompt)
        else:
            search_dict = search_prompt
    except json.JSONDecodeError as e:
        print(f"Error parsing JSON: {e}")
        return []

    compatible_items = []

    # Only process the first row
    for index, row in df.iterrows():
        item_features = {}
        all_keys_found = True

        # Load Key_Value_Pairs directly as a dict
        if isinstance(row['Key_Value_Pairs'], dict):
            item_features = row['Key_Value_Pairs']
        elif isinstance(row['Key_Value_Pairs'], str):
            try:
                item_features = json.loads(row['Key_Value_Pairs'])
            except json.JSONDecodeError as e:
                print(f"Error parsing JSON for row {index}: {e}")
                continue


        # Check each key-value in search_prompt
        for search_key, search_value in search_dict.items():
            key_match = False
            value_match = False

            # Try to find a key match (substring match)
            for item_key, item_value in item_features.items():
                if search_key == item_key or search_key in item_key:
                    key_match = True

                    # Check if the value matches exactly or if search_value is a substring of item_value
                    if search_value == item_value or search_value in item_value:
                        value_match = True
                        break

            # If no key match, check if value is in 'Features'
            if not key_match:
                features = item_features.get('Features', '')
                if search_value in features:
                    value_match = True

            # If no value match, mark all_keys_found as False and break
            if not value_match:
                all_keys_found = False
                break

        # If all keys were found, append the row to compatible_items
        if all_keys_found:
            compatible_items.append(row)

        break

    print(f"Total compatible items found: {len(compatible_items)}")
    return compatible_items
````

````python
compatible_items = search_compatible_items(data, translated_features)
print("Compatible Items:", compatible_items)
````
